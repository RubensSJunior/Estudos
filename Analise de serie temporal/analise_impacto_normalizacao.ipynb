### Analise de dados

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
coca_cola = pd.read_csv("./data/Coca Cola Stock/Coca-Cola_stock_history.csv")
residencial_power = pd.read_csv("./data/Residential power usage/power_usage_2016_to_2020.csv")
sunspots = pd.read_csv("./data/Sunspots/Sunspots.csv")
```

```python
cl_coca_cola = coca_cola[["Date","Close"]]
cl_residencial_power = residencial_power[["StartDate","Value (kWh)"]]
cl_sunspots = sunspots[["Date","Monthly Mean Total Sunspot Number"]]
```

```python
col_names = ["date","target"]
cl_coca_cola.columns = col_names
cl_residencial_power.columns = col_names
cl_sunspots.columns = col_names
```

```python
cl_coca_cola["date"] = cl_coca_cola["date"].apply(lambda x:pd.to_datetime(str(x).split(" ")[0]))

cl_residencial_power["date"] = cl_residencial_power["date"].apply(lambda x:pd.to_datetime(str(x).split(" ")[0]))
cl_residencial_power = cl_residencial_power.groupby("date").sum().reset_index()

cl_sunspots["date"] = pd.to_datetime(cl_sunspots["date"])
```

```python
plt.figure(figsize=(18, 4))
sns.lineplot(x="date",y="target",data=cl_coca_cola, label='Fechamento Aes Coca Cola')
```

```python
plt.figure(figsize=(18, 4))
sns.lineplot(x="date",y="target",data=cl_residencial_power, label='Consumo diario de Energia')
```

```python
plt.figure(figsize=(18, 4))
sns.lineplot(x="date",y="target",data=cl_sunspots, label='Quantidade de Manchas Solares')
```
